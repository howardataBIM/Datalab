# 3. Import mart database

```python
mart = pd.read_excel("excel/supermarket_sales.xlsx")
mart.head()
```

# 4.Create a basic BARPLOT

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total', data=mart)
```

# 5.Add HUE to the BARPLOT

```python
#add dimmension to graph
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total', hue='Gender', data=mart)
```

# 6. Make the barplot HORIZONTAL

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Total',y='Product line',data=mart)
```

# 7. Plot the bars in a given order

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',hue='Gender', data=mart,order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'],hue_order=['Male','Female'])
```

# 8. Add CAP on the ERROR BAR

```python
#Anadimos la linea horizontal a la barra de errores
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',hue='Gender', data=mart,order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'],hue_order=['Male','Female'],capsize=0.1)
```

# 9. Remove the error bar using CI

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'],hue_order=['Male','Female'],errorbar=None)
```

# 10.Change bar colors using COLOR attribute

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'],hue_order=['Male','Female'],color='black')
```

# 11.Using PALETTE attribute

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',hue=('Gender'),data=mart,order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'],hue_order=['Male','Female'],palette='cividis')
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',data=mart,order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'],palette='plasma')
```

# 12.Using SATURATION parameter

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',data=mart,order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'],palette='plasma'
           ,saturation=10)
```

# 13. change default aggregation method using estimator parameter

```python
#parametro suma
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',data=mart,order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel']
           ,palette='plasma'
           ,estimator=sum)
```

```python
#parametro mediana
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',data=mart,order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel']
           ,palette='plasma'
           ,estimator=np.median)
```
