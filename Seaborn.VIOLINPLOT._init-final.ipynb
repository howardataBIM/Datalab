# 1.Import al the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2.Import the supermart data excel file

```python
mart=pd.read_excel("excel/supermarket_sales.xlsx")
mart.head()
```

# 3.Create a basic VIOLINPLOT

```python
sns.violinplot(y='Total',data=mart)
```

# 4.Create a VIOLINPLOT on two categorical and one numeric variable and use split

```python
sns.violinplot(x='Payment',y='Total',data=mart)
```

```python
sns.violinplot(x='Payment',y='Total',hue='Gender', data=mart)
```

```python
sns.violinplot(x='Payment',y='Total',hue='Gender',data=mart,split=True)
```

```python
sns.violinplot(x='Payment',y='Total',hue='Gender',data=mart,split=True,inner='quartile')
```

# 5.Change the box in the VIOLINPLOT to horizontal lines

```python
sns.violinplot(x='Total',y='Payment',data=mart)
```

```python
sns.violinplot(x='Total',y='Payment',hue='Gender',data=mart)
```

# 6.Draw line for each observation in a VIOLINPLOT

```python
sns.violinplot(x='Payment',y='Total',hue='Gender',data=mart,split=True,inner='stick')
```

# 7.Change the amount of smoothing using bw attribute

```python
sns.violinplot(x='Payment',y='Total',hue='Gender',data=mart,split=True,inner='quartile',bw=.1)
```

# 8.Cut out the extreme values

```python
sns.violinplot(x='Payment',y='Total',hue='Gender',data=mart,split=True,inner='quartile',bw=.2,cut=0)
```
